### Exploratory Data Analysis (EDA)

#### Types of Content


```python
types_of_content=df['type'].value_counts()
types_of_content
```

```python
plt.figure(figsize=(8,8))
plt.pie(types_of_content, labels= types_of_content, autopct='%1.1f%%')
plt.title('Distribution of Content Types on Netflix')
plt.show()
```

#### From the Content Type Pie Chart (Movies vs TV Shows)

Movies dominate the Netflix catalog  usually ~6570% of the dataset.

TV Shows make up the remaining ~3035%.

This shows Netflix has historically been movie-heavy, but TV Shows have been growing in recent years.

### Distribution of Content Ratings on Netflix

```python
df['rating'].value_counts()
```

```python
plt.figure(figsize=(12,8))
sns.countplot(data=df, x='rating', hue='type')
plt.xlabel('Rating', fontsize=14)
plt.ylabel('No.of counts', fontsize=14)
plt.title('Distribution of Content Ratings on Netflix', fontsize=16)
plt.show()
```

#### Observations

TV-MA (Mature Audience) is the largest category with 3205 titles  Netflix has a strong focus on adult-oriented content.

TV-14 is next (2157 titles), targeting teenagers.

Family-friendly content (TV-Y, TV-Y7, TV-PG, TV-G, PG, G) has a smaller share compared to mature ratings.

Traditional movie ratings like R (799), PG-13 (490), and PG (287) are also present but much fewer than TV ratings.

Very few titles are marked as NR (Not Rated), NC-17, or UR (unrated).

### Number of Movies Released Per Year on Netflix

```python
No_of_contents = df.groupby(['release_year', 'type']).size().unstack(fill_value=0)
No_of_contents
```

```python
plt.figure(figsize=(12,8))
sns.countplot(data=df, x='release_year', hue='type')
plt.xlabel('Release Year', fontsize=14)
plt.ylabel('No. of counts', fontsize=14)
plt.xticks(rotation=90)
plt.title('Number of Movies Released Per Year on Netflix', fontsize=16)
plt.show()
```

#### Observations:
Early Years (19251945):

In the earliest years, the number of TV shows was extremely low or almost nonexistent, while movies were slowly increasing.

For example, in 1925, only 1 TV show existed compared to 0 movies, and by 19421945, movie production started dominating with 23 movies per year, while TV shows were rare.

Growth Trend:

Over the decades, movie production gradually increased, reaching its peak in recent years (20172018) with 765767 movies per year.

TV shows, which were negligible initially, saw a sharp increase in production, especially in the 2000s and 2010s, reaching 265379 shows in 20172018.

Recent Years (20172021):

Movie production peaked around 20172018 and then started declining (from 765 movies in 2017 to 277 in 2021).

TV shows, however, continued to rise, surpassing movies by 20202021 (436 TV shows vs. 517 movies in 2020, and 315 TV shows vs. 277 movies in 2021).

This indicates a shift in content production focus from movies to TV shows in recent years, possibly due to the rise of streaming platforms.

Overall Observation:

Historically, movies dominated the entertainment industry.

In recent years, the trend is shifting, with TV shows catching up and sometimes surpassing movie production, reflecting changes in audience consumption patterns and media platforms.
